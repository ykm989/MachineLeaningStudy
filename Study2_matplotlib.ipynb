# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

필요한 데이터 준비

```python
x = np.arange(0,6,0.1)
y = np.sin(x)

print(x)
print(y)
```

그래프 그리기  
plot 메서드를 이용하여 그래프를 그리고  
show 메서드를 이용하여 화면에 출력합니다.

```python
plt.plot(x,y)
```

```python
plt.show()
```

그래프 여러개를 한번에 그릴 수 있습니다.

```python
y2 = np.cos(x)

plt.plot(x,y)
plt.plot(x,y2, linestyle="--", label ="two")
plt.xlabel('x name')
plt.ylabel('y name')
plt.title('title')
plt.legend()
plt.show()
```

matplotlib에는 이미지를 그리는 메서드도 존재
